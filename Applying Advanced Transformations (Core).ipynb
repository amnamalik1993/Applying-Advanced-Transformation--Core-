# Data Transformation

## First Data Set

```python
info_df['Hero|Publisher'].head(2)
```

```python
info_df['Hero|Publisher'].str.split('|',expand=True)
```

```python
info_df[['Hero','Publisher']] = info_df['Hero|Publisher'].str.split('|',expand=True)
info_df.head(2)
```

```python
## drop the original column 
info_df = info_df.drop(columns=['Hero|Publisher'])
info_df.head(2)
```

```python
## single value from the Measurements column
measure = info_df.loc[0,"Measurements"]
print(type(measure))
measure
```

```python
measure = measure.replace("'",'"')
measure
```

```python
## Lets use json.loads
fixed_measure = json.loads(measure)
print(type(fixed_measure))
fixed_measure
```

```python
## use .str.replace to replace all single quotes
info_df['Measurements'] = info_df['Measurements'].str.replace("'",'"')
## Apply the json.loads to the full column
info_df['Measurements'] = info_df['Measurements'].apply(json.loads)
info_df['Measurements'].head()
```

```python
## check a single value after transformation
test_measure = info_df.loc[0, 'Measurements']
print(type(test_measure))
test_measure
```

```python
height_weight = info_df['Measurements'].apply(pd.Series)
height_weight
```

```python
# concat hgt_wht with original dataframe
info_df = pd.concat((info_df, height_weight), axis = 1)
info_df.head(2)
```

```python
# Dropping the Measurements column
info_df = info_df.drop(columns=['Measurements'])
info_df.head(2)
```

```python
# Make a list of all characters to replace
to_replace = [' ', 'cm']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    info_df['Height'] = info_df['Height'].str.replace(char,'',regex=False)
    
info_df['Height'].head()
```

```python
# Changing Height to float
info_df['Height'] = info_df['Height'].astype(float)
```

```python
# Make a list of all characters to replace
to_replace = [' ', 'kg']
# run a loop to replace all of the characters in the list at once
for char in to_replace:
    info_df['Weight'] = info_df['Weight'].str.replace(char,'',regex=False)
    
info_df['Weight'].head()
```

```python
# Changing Weight to float
info_df['Weight'] = info_df['Weight'].astype(float)
```

```python
info_df.info()
```

## Second Data Set

```python
power_df.head()
```

```python
power_df['powers_split'] = power_df['Powers'].str.split(",")
power_df.head()
```

```python
exploded = power_df.explode('powers_split')
```

```python
col_to_make = exploded['powers_split'].dropna().unique()
col_to_make
```

```python
for col in col_to_make:
    power_df[col] = power_df['Powers'].str.contains(col)
power_df.head()
```

```python
# Dropping the Powers and powers_split columns
power_df = power_df.drop(columns=['Powers','powers_split'])
power_df.head()
```

# Putting Columns Together

```python
merged = pd.merge(info_df, power_df, left_on = 'Hero', right_on = 'hero_names')

merged.head()
```

```python
# 1: Compare the average weight of super powers who have Super Speed to those who do not.

merged.groupby('Super Speed')['Weight'].mean()
```

```python
# 2: What is the average height of heroes for each publisher?

merged.groupby('Publisher')['Height'].mean()
```
